```python
# dashdott-->

plt.plot(batter['index'],batter['V Kohli'],color='green',linestyle='dashdot')
plt.plot(batter['index'],batter['RG Sharma'],color='red',linestyle='dashdot')
plt.title('V kohli vs rohit sharma career caparasion' )
plt.xlabel('years')
plt.ylabel('runs')
plt.show()
```

```python
# width of line -->
plt.plot(batter['index'],batter['V Kohli'],color='green',linestyle='dashdot',linewidth='3')
plt.plot(batter['index'],batter['RG Sharma'],color='red',linestyle='dashdot',linewidth='2')
plt.title('V kohli vs rohit sharma career caparasion' )
plt.xlabel('years')
plt.ylabel('runs')
plt.show()
```

```python
# marker size
plt.plot(batter['index'],batter['V Kohli'],color='green',linestyle='dashdot',marker='D',markersize=10)
plt.plot(batter['index'],batter['RG Sharma'],color='red',linestyle='dashdot',marker='o')
plt.title('V kohli vs rohit sharma career caparasion' )
plt.xlabel('years')
plt.ylabel('runs')
plt.show()
```

```python
# legend()--locqtiom
plt.plot(batter['index'],batter['V Kohli'],color='green',linestyle='dashdot',marker='D',markersize=10,label='V Kohli')
plt.plot(batter['index'],batter['RG Sharma'],color='red',linestyle='dashdot',marker='o',label='RG Sharma')
plt.title('V kohli vs rohit sharma career caparasion' )
plt.xlabel('years')
plt.ylabel('runs')


plt.legend()
```

```python
# simple Bar charts
children=[10,20,30,50,40]
color=['red','green','blue','yellow','orange']
plt.bar(color,children)
```

```python
# bar chart using data
df=pd.read_csv('/content/batsman_season_record.csv')
df
```

```python
plt.bar(df['batsman'],df['2015'],width=0.5)
```

```python
# multiple bar charts  -- jugaad
plt.bar(np.arange(df.shape[0])- 0.2,df['2015'],width=0.2,color='red')
plt.bar(np.arange(df.shape[0]),df['2016'],width=0.2,color='yellow')
plt.bar(np.arange(df.shape[0])+ 0.2,df['2017'],width=0.2,color='blue')
plt.xticks(np.arange(df.shape[0]),df['batsman'])
plt.show()
```

```python
#problem-> children=[10,20,30,50,40]
color=['red red red red red red','green green green','blue blue blue ','yellow yellow yellow yellow','orange orange orange']
plt.bar(color,children)
```

```python
# over lap - problem
```

```python
# solution -->
plt.bar(color,children,color='purple')
plt.xticks(rotation='vertical')
```

```python
# stacked Bar chart
plt.bar(df['batsman'],df['2017'],label='2017')
plt.bar(df['batsman'],df['2016'],bottom=df['2017'],label='2016')
plt.bar(df['batsman'],df['2015'],bottom=df['2016'] + df['2017'],label='2015')
plt.legend()
```

# Pie Chart

```python
# simple example
p=[10,50,30,60,100,]
subject=['math','science','social','hindi','english']
```

```python
plt.pie(p,labels = subject)
plt.show()
# problem - kuch samajh nhi aa rha hai
```

```python
# labels-
plt.pie(p,labels=subject)
plt.show()
```

```python
# datasets
df=pd.read_csv('/content/gayle-175.csv')
df
plt.pie(df['batsman_runs'],labels=df['batsman'])
plt.show()
```

```python
# percentage and colors
df=pd.read_csv('/content/gayle-175.csv')
df
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1fg%%',colors=['blue','skyblue','green','red','yellow','purple'])
plt.savefig('sample.png')
```

```python
# Extract a pie --> explode
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1fg%%',explode=[0.1,0,0,0,0,0])
plt.show()
```

```python
# shadow
plt.pie(df['batsman_runs'],labels=df['batsman'],autopct='%0.1fg%%',shadow=True)
plt.show()
```
